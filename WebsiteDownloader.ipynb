```
import requests

url = 'http://is.gd/jcMvjB'
r = requests.get(url)
url = r.url
print(url)
```

```
from pywebcopy import save_webpage

name = url.split('https://')[1]
name = name.split('.')[0]
download_folder = '/content/webpages'    
kwargs = {'bypass_robots': True, 'project_name': name}
save_webpage(url, download_folder)
```

```
import requests

content = requests.get(url).content

with open('/content/webpages/medium.com/index.html', 'wb') as f:
  f.write(content)
```
