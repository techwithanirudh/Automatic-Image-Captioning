```
import language_check 

tool = language_check.LanguageTool('en-US') 
line = 'i is an student.'
matches = tool.check(line) 
for mistake in matches: 
  mistake = str(mistake)
  mistake = mistake.split('Message: ')[1].replace('^', '')
  # mistake = mistake.split('})]')[1]
  print(mistake) 
```
